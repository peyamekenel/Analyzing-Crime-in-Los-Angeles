| Column     | Description              |

| `'AREA NAME'` | The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. | `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |

| `'LOCATION'` | Street address of the crime. |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"],dtype={"TIME OCC": str})
crimes.head()
```

```python
# Start coding here
# Use as many cells as you need
```

```python
crimes["Hour"]=crimes["TIME OCC"].astype(int)//100
```

```python
hourly_crime_counts=crimes["Hour"].value_counts()
```

```python
peak_crime_hour=hourly_crime_counts.idxmax()
```

```python
peak_crime_hour
```

```python
sns.barplot(x=hourly_crime_counts.index,y=hourly_crime_counts.values)
```

```python
night_crimes = crimes[(crimes['Hour'] >= 22) | (crimes['Hour'] <= 3)]
```

```python
peak_night_crime_location = night_crimes['AREA NAME'].value_counts().idxmax()
```

```python
peak_night_crime_location
```

```python
age_bins = [0, 17, 25, 34, 44, 54, 64, 100]
```

```python
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
```

```python
crimes["Age Group"]=pd.cut(crimes["Vict Age"],bins=age_bins,labels=age_labels)
```

```python
victim_ages = crimes['Age Group'].value_counts().reindex(age_labels, fill_value=0)
```
